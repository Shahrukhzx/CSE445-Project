# Handling Mixed dtypes

```python
# Converting the TotalPatientsInEmergency column to numeric
dataset['TotalPatientsInEmergency'] = pd.to_numeric(dataset['TotalPatientsInEmergency'], errors='coerce')
dataset.info()
```

# Handling the missing values

```python
dataset['EducationTitle'].unique()
```

```python
check=(dataset["Gender"] == '0').sum()
check
```

```python
#Dropping the uncessary columns
dataset.drop(['EcYear','EcNumber','EmergencyArea','Gender','HospitalName','Reason','EducationTitle'],axis=1,inplace=True)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
# using forwardfill to impute the CallTime column
dataset['CallTime'].fillna(method='ffill', inplace=True)
```

```python
dataset.isnull().sum()
```

```python
#Dropping the null rows
dataset=dataset.dropna()
dataset.isnull().sum()
```

# Encoding Categorical Variables

```python
dataset["Gender"].unique()
```

```python
# LabelEncodding for PatientStatus
status_mapping = {
    'Alive & stable': 0,
    'Alive & unstable': 1,  
    'Dead': 2  
}

# Applying the mapping to the 'PatientStatus' column
dataset['PatientStatus'] = dataset['PatientStatus'].map(status_mapping)
```

```python
# Show the first few rows to verify the encoding
dataset.head()
```

```python
# Target-encodding for InjuryType and Cause
def target_encoding(column, target):
    return dataset.groupby(column)[target].transform('mean')
dataset['InjuryTypeEncoded'] = target_encoding('InjuryType', 'PatientStatus')
dataset['CauseEncoded'] = target_encoding('Cause', 'PatientStatus')
```

# Generating a heatmap

```python
# creating a new df for the numeric data's only for generating a heat map
numeric_dataset = dataset.drop(['InjuryType','Cause'],axis=1,inplace=False)
```

```python
numeric_dataset[:10]
```

```python
# Compute the correlation matrix
corr_matrix = numeric_dataset.corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Road Accident Dataset')
plt.show()
```

```python
# Scaling

# List the numeric columns that you want to normalize
numeric_columns = [
    'TotalPatientsInEmergency', 'Age', 'responsetime', 
]

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()

# Apply the scaler to the numeric columns
numeric_dataset[numeric_columns] = scaler.fit_transform(dataset[numeric_columns])

# Now your dataset is normalized
print(numeric_dataset.head())
```
